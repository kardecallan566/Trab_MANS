## Tentativas

```python
Machine = ["M1_P1", "M1_P2", "M1_P3", "M1_P4", "M2_P1", "M2_P2", "M2_P3", "M3_P1", "M3_P2", "M3_P3"]
```

```python
for i in Machine:
    print(f"df_permutations[{i}]")
```

```python
Inicio = [38, 33, 36, 20, 32, 29, 46, 38, 34, 40]
permutation_list = list(itertools.permutations(Inicio))
```

```python
x1 = 38+29+36+20
x2 = 38+33+46
x3 = 32+34+40

max(x1, x2, x3)
```

```python
df_permutations = pd.DataFrame(permutation_list, columns=Machine)
df_permutations["Maq1"] = df_permutations[["M1_P1", "M1_P2", "M1_P3", "M1_P4"]].sum(axis=1)
df_permutations["Maq2"] = df_permutations[["M1_P1", "M2_P2", "M2_P3"]].sum(axis=1)
df_permutations["Maq3"] = df_permutations[["M3_P1", "M3_P2", "M3_P3"]].sum(axis=1)
df_permutations["MaquinaDemora"] = df_permutations[["Maq1", "Maq2", "Maq3"]].max(axis=1)

# df_permutations.drop(["Maq1", "Maq2", "Maq3"], axis=1, inplace=True)
```
