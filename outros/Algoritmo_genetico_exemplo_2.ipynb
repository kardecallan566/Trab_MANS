```python
from itertools import permutations
import pandas as pd
import random
import numpy as np


def cost_calculation(df_input, distance_matrix, trabalhadores):
    for jj, r in df_input.iterrows():
        cost = 0
        perm=r['permutacao']
        for kk in trabalhadores:
            cost += distance_matrix.loc[perm[kk-1], kk]

        df_input.loc[jj, 'custo'] = cost

    return df_input


distance_matrix = pd.DataFrame([[20, 25, 17, 13],
                                [15, 30, 19, 21],
                                [12, 24, 27, 32],
                                [17, 33, 23, 28]], columns=[1, 2, 3, 4])
distance_matrix.index += 1

trabalhadores = list(range(1, 5))
# Geração das permutações da lista [1,2,3,4]
permutation_list = list(permutations(trabalhadores))
print(permutation_list)

# Criação de uma tabela, em que cada linha representa um cromossoma/ solução para o problema
nperms=4*3*2 #Número de todas as permutações possíveis
df_permutations = pd.DataFrame(columns=['permutacao','custo'])
for i in range(0,nperms,1):
    perm=permutation_list[i]
    vec=[None]*len(trabalhadores)
    for j in range(0,len(trabalhadores),1):
        vec[j]=perm[j]
    df_permutations.loc[i,'permutacao']=vec
print(df_permutations)
```

# Geração da População Inicial e Cálculo do Valor de Aptidão de cada cromossoma

```python
# A função cost_calculation calcula o custo de cada cromossoma/ solução

def cost_calculation(df_input, distance_matrix, trabalhadores):
    for jj, r in df_input.iterrows():
        cost = 0
        perm=r['permutacao']
        for kk in trabalhadores:
            cost += distance_matrix.loc[perm[kk-1], kk]

        df_input.loc[jj, 'custo'] = cost

    return df_input


distance_matrix = pd.DataFrame([[20, 25, 17, 13],
                                [15, 30, 19, 21],
                                [12, 24, 27, 32],
                                [17, 33, 23, 28]], columns=[1, 2, 3, 4])
distance_matrix.index += 1
# número de cromossomas da população
n_cromossomas = 20

randomlist = random.sample(range(0, len(permutation_list)), n_cromossomas)
print(randomlist)
df_inicial = df_permutations.loc[randomlist]
print('População Inicial')
print(df_inicial)
print()
print('População Inicial (na última coluna tem o custo do cromossoma)')
df_inicial = cost_calculation(df_inicial, distance_matrix, trabalhadores)
print(df_inicial)
```

# Seleção dos Pais através do Processo de Seleção por Torneio

```python
# k representa o número de cromossomas selecionados da população de forma aleatória
k = 5

# Seleção do Pai1: dos k cromossomas selecionados, o Pai1 corresponde ao cromossoma de melhor aptidão (i.e., de menor custo)
pai_list1 = random.sample(list(df_inicial.index), k)
print(pai_list1)
pai1 = df_inicial.loc[pai_list1, 'custo'].astype(int).idxmin() #aqui, .astype(int) indica ao python que o custo é uma variável do tipo inteiro (por algum motivo existe ambiguidade)

# Seleção do Pai2: dos k cromossomas selecionados, o Pai2 corresponde ao cromossoma de melhor aptidão, mas garantindo que é 
# diferente do Pai1
abc = True
while abc:
    pai_list2 = random.sample(list(df_inicial.index), k)
    print(pai_list2)
    pai2 = df_inicial.loc[pai_list2, 'custo'].astype(int).idxmin() #ver nota do mesmo comando do pai1
    if pai1 != pai2:
        abc = False
print('Pai1:',pai1) 
print('Pai2:',pai2)        
```

# Operação de crossover: Crossover a um ponto

```python
max_distance = distance_matrix.max().max()

# Geração aleatória do número m
crossover_options = list(range(1, len(trabalhadores) - 1))
m = random.choice(crossover_options)
print('m:',m)

df_crossover = df_inicial.loc[[pai1, pai2], :]

# Geração dos cromossomas filho
p1=df_crossover.loc[pai1,'permutacao'].copy()
p2=df_crossover.loc[pai2,'permutacao'].copy()
filho1=p1.copy()
filho2=p2.copy()
for i in range(m,len(trabalhadores),1):
    filho1[i]=p2[i]
    filho2[i]=p1[i]
df_crossover.loc['filho1','permutacao']=filho1
df_crossover.loc['filho2','permutacao']=filho2

df_crossover = cost_calculation(df_crossover, distance_matrix, trabalhadores)

#print('Cromossoma Filho1(após crossover):')
#print(df_crossover.loc['filho1',:])
#print()
#print('Cromossoma Filho2(após crossover):')
#print(df_crossover.loc['filho2',:])
#print()

print('Cromossomas filhos após crossover:')
print(df_crossover)

no_changes=False
# Averiguação da admissibilidade dos cromossomas filho
# Filho 1
feasible=True
contador=np.zeros(len(trabalhadores))
for i in range(0,len(trabalhadores),1):
    contador[filho1[i]-1]+=1
    if contador[filho1[i]-1]>1.0:
        feasible=False

if feasible==False:
    print('Filho 1 não admissivel')
    no_changes=True

    filho1_a=filho1.copy()
    #Algoritmo para reparar a admissibilidade da solução: para cada elemento que aparece mais do que uma vez na solução, substituir
    #a última repetição pelo primeiro elemento que não aparece na solução (versão adaptada para permitir que haja mais do que uma repetição)
    finished=False
    while finished==False:
        finished=True
        for i in range(0,len(trabalhadores),1):
            if contador[i]>1:
                #Procuramos a última entrada em que i aparece
                last=0
                for k in range(0,len(trabalhadores),1):
                    if filho1_a[len(trabalhadores)-1-k]==i+1:
                        last=len(trabalhadores)-1-k
                        break
                #Procuramos o primeiro elemento que não aparece
                not_included=0
                for k in range(0,len(trabalhadores),1):
                    if contador[k]<1:
                        not_included=k+1
                        break
                #Fazemos a substituição e atualizamos os contadores
                filho1_a[last]=not_included
                contador[not_included-1]+=1
                contador[i]-=1
                finished=False
                break
    
    
    df_crossover.loc['filho1_admissivel', 'permutacao'] = filho1_a
    
    df_crossover = cost_calculation(df_crossover, distance_matrix, trabalhadores)

    df_crossover.loc['filho1_penalty', :] = df_crossover.loc['filho1', :].copy()
    df_crossover.loc['filho1_penalty', 'custo'] += 2 * max_distance
else:
    df_crossover.loc['filho1_admissivel', :] = df_crossover.loc['filho1', :].copy()

# Filho 2
feasible=True
contador=np.zeros(len(trabalhadores))
for i in range(0,len(trabalhadores),1):
    contador[filho2[i]-1]+=1
    if contador[filho2[i]-1]>1.0:
        feasible=False

if feasible==False:
    print('Filho 2 não admissivel')
    no_changes=True

    filho2_a=filho2.copy()
    #Algoritmo para reparar a admissibilidade da solução: para cada elemento que aparece mais do que uma vez na solução, substituir
    #a última repetição pelo primeiro elemento que não aparece na solução (versão adaptada para permitir que haja mais do que uma repetição)
    finished=False
    while finished==False:
        finished=True
        for i in range(0,len(trabalhadores),1):
            if contador[i]>1:
                #Procuramos a última entrada em que i aparece
                last=0
                for k in range(0,len(trabalhadores),1):
                    if filho2_a[len(trabalhadores)-1-k]==i+1:
                        last=len(trabalhadores)-1-k
                        break
                #Procuramos o primeiro elemento que não aparece
                not_included=0
                for k in range(0,len(trabalhadores),1):
                    if contador[k]<1:
                        not_included=k+1
                        break
                #Fazemos a substituição e atualizamos os contadores
                filho2_a[last]=not_included
                contador[not_included-1]+=1
                contador[i]-=1
                finished=False
                break
    
    
    df_crossover.loc['filho2_admissivel', 'permutacao'] = filho2_a
    
    df_crossover = cost_calculation(df_crossover, distance_matrix, trabalhadores)

    df_crossover.loc['filho2_penalty', :] = df_crossover.loc['filho2', :].copy()
    df_crossover.loc['filho2_penalty', 'custo'] += 2 * max_distance    
else:
    df_crossover.loc['filho2_admissivel', :] = df_crossover.loc['filho2', :].copy()
    
print('Cromossomas filhos (com correção de admissibilidade e com penalização):')
print(df_crossover)

#Na próxima população os pais que foram escolhidos nesta iteração são substituídos pelos filhos gerados. O resto da população mantém-se
df_nextgen = df_inicial.copy()
df_nextgen.loc[pai1, :] = df_crossover.loc['filho1_admissivel', :]
df_nextgen.loc[pai2, :] = df_crossover.loc['filho2_admissivel', :]

print('Nova população:')
print(df_nextgen)
```

# Operação Mutação: Mutação por Troca

```python
#mutação filho 1
rand_1 = random.random()
print('rand_1:',rand_1)
if rand_1 <= 0.1:
    print('Mutação - filho1')
    mutants = random.sample(trabalhadores, 2)
    p1=df_nextgen.loc[pai1,'permutacao'].copy()
    auxiliar = p1[mutants[0]-1]
    p1[mutants[0]-1] = p1[mutants[1]-1]
    p1[mutants[1]-1] = auxiliar
    df_nextgen.loc[pai1,'permutacao']=p1.copy()

    df_nextgen = cost_calculation(df_nextgen, distance_matrix, trabalhadores)
    print(df_nextgen.loc[pai1,:])

#mutação filho 2
rand_2 = random.random()
print('rand_2:',rand_2)
if rand_2 <= 0.1:
    print('Mutação - filho2')
    mutants = random.sample(trabalhadores, 2)
    p2=df_nextgen.loc[pai2,'permutacao'].copy()
    auxiliar = p2[mutants[0]-1]
    p2[mutants[0]-1] = p2[mutants[1]-1]
    p2[mutants[1]-1] = auxiliar
    df_nextgen.loc[pai2,'permutacao']=p2.copy()


    df_nextgen = cost_calculation(df_nextgen, distance_matrix, trabalhadores)
    
    print(df_nextgen.loc[pai2,:])
```
